```python
#Offensive
df['GoaltoShotPercentage']=df['GoalsScored']/df['ShotsonTarget']
df['TurnoversPerPassAttempt']= (df['Offside']+df['OutoBounds']+df['Intercepted']+df['Blocked'])/df['PassAtt']
df['GoalCreationActionPerShotCreationAction']= df['GoalCreationAction']/df['ShotCreationAction']
df['PenaltyConversionRate']=df['PKConverted']/df['PKWon']

#Defensive
df['GoaltoShotAgainstPercentage']=df['GoalsAllowed']/df['ShotsonTargetAgainst']
df['PressSuccessRate']= df['PressSuccess']/df['PressAttempt']
df['TackleandDribbleStopPercent']= (df['TacklesWon']+df['VsDribblerWon'])/(df['TacklesAttempted']+ df['VsDribblerAttempted'])
df['PenaltySaveRate']=df['PKsaved']/df['PKattempt']
```

```python
newdf= df[['Squad','W','GoaltoShotPercentage',
           'TurnoversPerPassAttempt','GoalCreationActionPerShotCreationAction','PenaltyConversionRate',
           'GoaltoShotAgainstPercentage','PressSuccessRate','TackleandDribbleStopPercent','PenaltySaveRate']]
newdf.loc['mean'] = newdf.mean()
newdf
```

# 2- Regression

```python
newdf= df[['GoaltoShotPercentage',
           'TurnoversPerPassAttempt','GoalCreationActionPerShotCreationAction','PenaltyConversionRate',
           'GoaltoShotAgainstPercentage','PressSuccessRate','TackleandDribbleStopPercent','PenaltySaveRate']]
# Correlation heatmap
newdf.corr()
```

```python
newdf['GSP-OGSP']=newdf['GoaltoShotPercentage']-newdf['GoaltoShotAgainstPercentage']
newdf['TPP-DTPP']=newdf['TurnoversPerPassAttempt']-newdf['PressSuccessRate']
newdf['GSCA-OGSCA']=newdf['GoalCreationActionPerShotCreationAction']-newdf['TackleandDribbleStopPercent']
newdf['PCR-PSR']=newdf['PenaltyConversionRate']-newdf['PenaltySaveRate']
Regdata=newdf[['GSP-OGSP','TPP-DTPP','GSCA-OGSCA','PCR-PSR']]
Regdata1=newdfrank[['W']]
FinalData= Regdata1.join(Regdata).drop(index='mean')
```

```python
FinalData
```

```python
### splitting the testing and training data
train_set, test_set = train_test_split(FinalData, test_size = 0.2, random_state = 1234)
```

```python
X_train= train_set.drop('W', axis = 1)

y_train=train_set['W']

X_test= test_set.drop('W', axis = 1)

y_test= test_set['W']

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)  
```

```python
lr=LinearRegression()
lr.fit(X_train, y_train)
lr_pred= lr.predict(X_test)

print(lr.coef_)
print(lr.intercept_)

print(explained_variance_score(y_test,lr_pred)) 
print(r2_score(y_test,lr_pred)) 
print(mean_squared_error(y_test,lr_pred))
```

```python
mod = sm.OLS(y_train,X_train)

fii = mod.fit()

print(fii.summary())
```
